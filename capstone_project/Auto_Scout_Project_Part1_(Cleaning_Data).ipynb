```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import skimpy

import warnings 
warnings.filterwarnings('ignore')

sns.set_style("darkgrid")
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 1000)
```

```python
## Reading Data

df0 = pd.read_json("as24_cars.json")
df = df0.copy()
```

```python
df.sample(5)
```

```python
## Checking null values 

df.isnull().sum() / len(df) * 100
```

```python
## Cleaning the columns

df.columns = skimpy.clean_columns(df).columns
```

```python
## Clean Make column

df.make = df.make.str.strip('\n')
```

```python
df.make.head()
```

```python
## Clean Model column

df.model = df.model.map(lambda x: ",".join(x) if type(x) == list else x).str.strip('\n, ')
```

```python
df.loc[df[(df.model == '') & (df.make == 'Opel')].index, 'model'] = 'Excellence'
```

```python
## Clean Price column

df.price = df.price.str.strip('€.-').str.replace(',', '').astype('float')
```

```python
df.price.head()
```

```python
## Clean Body Type column

df.body_type = df.body_type.map(lambda x: ",".join(x) if type(x) == list else x).str.strip('\n, ').str.replace(',','')
```

```python
df.body_type.head()
```

```python
## Clean Type column

df.type = df.type.map(lambda x: ",".join(x) if type(x) == list else x).str.strip('\n, ').str.replace(',','')
```

```python
df.type.head()
```

```python
## Clean Doors column

df.doors = df.doors.map(lambda x: ",".join(x) if type(x) == list else x).str.strip('\n, ').str.replace(',','').astype('float')
```

```python
df.doors.head()
```

```python
## Clean Country Version column

df.country_version = df.country_version.map(lambda x: ",".join(x) if type(x) == list else x).str.strip('\n, ').str.replace(',','')
```

```python
df.country_version.head()
```

```python
## Clean Offer Number column

df.offer_number = df.offer_number.map(lambda x: ",".join(x) if type(x) == list else x).str.strip('\n, ').str.replace(',','')
```

```python
df.offer_number.head()
```

```python
## Clean Warranty column

df.warranty = df.warranty.map(lambda x: ",".join(x) if type(x) == list else x).str.strip('\n, ').str.replace(',','')
```

```python
df.warranty.head()
```

```python
## Clean Mileage column

df.mileage = df.mileage.str.split(" ").map(lambda x: x[0] if type(x) == list else x).str.replace(',', '').astype('float')
```

```python
## Clean First Registration column, Create Age column

df["age"] = 2022 - df.first_registration.str.split('/').map(lambda x: x[1] if type(x) == list else x).astype('float')
```

```python
df.age.head()
```

```python
## Clean Gearbox column

df.gearbox = df.gearbox.map(lambda x: ",".join(x) if type(x) == list else x).str.strip('\n, ')
```

```python
df.gearbox.head()
```

```python
## Clean Desc column

df.desc = df.desc.map(lambda x: ",".join(x) if type(x) == list else x)
```

```python
## Clean Seats column

df.seats = df.seats.map(lambda x: ",".join(x) if type(x) == list else x).str.strip('\n, ').str.replace(',','')
```

```python
df.seats = df.seats.astype('float')
```

```python
df.seats.head()
```

```python
## Clean Power column, create new columns

df['power_kW'] = df.power.map(lambda x: ",".join(x) if type(x) == list else x).str.extract('(\d+) kW').astype('float')
```

```python
df.power_kW.head()
```

```python
df['horse_power'] = df.power.map(lambda x: ",".join(x) if type(x) == list else x).str.extract('\((\d+) hp\)').astype('float')
```

```python
df.horse_power.head()
```

```python
## Clean Engine Size column

df.engine_size = (df.engine_size.map(lambda x: x[0].strip("\n c").replace(",", "") if type(x) == list else x).astype(float) / 1000)\
        .astype(str).str.extract("(\d+\.*\d)").astype(float)
```

```python
df.engine_size.head()
```

```python
## Clean Gears column

df.gears = df.gears.map(lambda x: ",".join(x) if type(x) == list else x).str.strip('\n, ')
```

```python
df.gears = df.gears.astype('float')
```

```python
df.gears.head()
```

```python
## Clean Co2 Emissions column

df.co_emissions = df0["CO₂-emissions"].str.extract('(\d+) ').astype('float')
```

```python
df.co_emissions.sample(5)
```

```python
## Clean Manufacturer Colour column

df.manufacturer_colour = df.manufacturer_colour.str.title().str.strip(' ')
```

```python
df.manufacturer_colour.head()
```

```python
## Clean Drivetrain column

df.drivetrain = df.drivetrain.map(lambda x: ",".join(x) if type(x) == list else x).str.strip('\n, ').str.replace(',', '')
```

```python
df.drivetrain.head()
```

```python
## Clean Cylinders column

df.cylinders = df.cylinders.map(lambda x: ",".join(x) if type(x) == list else x).str.strip('\n, ').astype('float')
```

```python
df.cylinders.head()
```

```python
## Clean Fuel Consumption column

import itertools

df["f_consumption_per_100_combined"] = df.fuel_consumption.map(lambda x: ",".join(list(itertools.chain.from_iterable(x))) if type(x) == list else x)\
            .str.extract('(.+) l/100 km \(comb.\)').astype('float')
```

```python
## Clean Comfort & Convenience column, create categories

def standardize(value):
    if value == np.nan:
        return np.nan
    if value <= 10:
        return 'Standard'
    if value > 10 and value <= 20:
        return 'Premium'
    if value > 20:
        return 'Deluxe'

df["comfort_&_convenience"] = df["comfort_&_convenience"].map(lambda x: ''.join(x) if type(x) == list else x)

df["comfort_convenience_pack"] = df["comfort_&_convenience"].map(lambda x: ''.join(x) if type(x) == list else x)\
        .map(lambda x: x.split(',') if type(x) == str else x)\
        .map(lambda x: len(x) if type(x) == list else x).map(lambda x: standardize(x) if type(x) == float else x)
```

```python
## Clean Safety & Security column, create categories

def standardize_2(value):
    if value == np.nan:
        return np.nan
    if value <= 12:
        return 'Standard'
    if value > 12 and value <= 24:
        return 'Premium'
    if value > 24:
        return 'Deluxe'

df['safety_&_security'] = df['safety_&_security'].map(lambda x: ''.join(x) if type(x) == list else x)

df['safety_security_pack'] = df['safety_&_security'].map(lambda x: ''.join(x) if type(x) == list else x)\
        .map(lambda x: x.split(',') if type(x) == str else x)\
        .map(lambda x: len(x) if type(x) == list else x).map(lambda x: standardize_2(x) if type(x) == float else x)
```

```python
## Clean Entertainment & Media column, create categories

def standardize_3(value):
    if value == np.nan:
        return np.nan
    if value <= 6:
        return 'Standard'
    if value > 6 and value <= 12:
        return 'Premium'
    if value > 12:
        return 'Deluxe'

df['entertainment_&_media'] = df['entertainment_&_media'].map(lambda x: ''.join(x) if type(x) == list else x)

df['entertainment_pack'] = df['entertainment_&_media'].map(lambda x: ''.join(x) if type(x) == list else x)\
        .map(lambda x: x.split(',') if type(x) == str else x)\
        .map(lambda x: len(x) if type(x) == list else x).map(lambda x: standardize_3(x) if type(x) == float else x)
```

```python
## Clean Extras column, create categories

def standardize_4(value):
    if value == np.nan:
        return np.nan
    if value <= 9:
        return 'Standard'
    if value > 9 and value <= 18:
        return 'Premium'
    if value > 18:
        return 'Deluxe'

df.extras = df.extras.map(lambda x: ''.join(x) if type(x) == list else x)

df['extra_pack'] = df.extras.map(lambda x: ''.join(x) if type(x) == list else x)\
        .map(lambda x: x.split(',') if type(x) == str else x)\
        .map(lambda x: len(x) if type(x) == list else x).map(lambda x: standardize_4(x) if type(x) == float else x)
```

```python
## Clean Empty Weight column

df.empty_weight = df.empty_weight.map(lambda x: ''.join(x) if type(x) == list else x).str.strip('\nkg ')\
            .str.replace(',','').astype('float')
```

```python
## Clean Model Code column

df.model_code = df.model_code.map(lambda x: ''.join(x) if type(x) == list else x).str.strip('\n, ')
```

```python
df.model_code.sample(5)
```

```python
## Clean Upholstery column

df.upholstery = df.upholstery.str.title().str.strip(' ')
```

```python
df.upholstery.sample(5)
```

```python
## Clean Previous Owner column
```

```python
df.previous_owner = df.previous_owner.map(lambda x: x[1] if type(x) == list else x).astype('float')
```
